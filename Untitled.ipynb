```julia
function rawlatexboundingbox(lstr::LaTeXString)
    sentence = generate_tex_elements(lstr)
    els = filter(x -> x[1] isa TeXChar, sentence)
    texchars = [x[1] for x in els]
    pos_x = [x[2][1] for x in els]
    pos_y = [x[2][2] for x in els]
    scales = [x[3] for x in els]
    textw = maximum(inkwidth.(texchars) .* scales .+ pos_x)
    bottom = bottominkbound.(texchars)
    bottom_pt = Point(0, maximum(-bottom .* scales .- pos_y))
    top_pt = Point(textw, -maximum(inkheight.(texchars) .* scales .+ pos_y .+ bottom))

    return (bottom_pt, top_pt)
end

# import MathTeXEngine: inkheight, inkwidth, bottominkbound
rawlatexboundingbox(L"\langle")
```

```julia
using MathTeXEngine
d = Drawing(200,200,:svg)
origin()
fontsize(18)
t = L"\sum^N_{n=1} \frac{x}{n}"

for halign in [:left,:right,:center], valign in [:baseline, :bottom, :top, :middle]
    text(t, halign=halign, valign=valign)
    @show round.(latextextsize(t);digits=2)== (37.78, 45.98)
    
    
    if halign === :left
        @show round(latexboundingbox(t, halign=halign, valign=valign)[1][1];digits=2) == 0.0
        @show round(latexboundingbox(t, halign=halign, valign=valign)[2][1];digits=2) == 37.78
    elseif halign === :right
        @show round(latexboundingbox(t, halign=halign, valign=valign)[1][1];digits=2) == -37.78
        @show round(latexboundingbox(t, halign=halign, valign=valign)[2][1];digits=2) == 0.0
    elseif halign === :center
        @show round(latexboundingbox(t, halign=halign, valign=valign)[1][1];digits=2) == -37.78/2
        @show round(latexboundingbox(t, halign=halign, valign=valign)[2][1];digits=2) == 37.78/2
    end
    
    if valign === :baseline || valign === :bottom 
        @show round(latexboundingbox(t, halign=halign, valign=valign)[1][2];digits=2) == 18.25
        @show round(latexboundingbox(t, halign=halign, valign=valign)[2][2];digits=2) == -27.72
    elseif valign === :top
        @show round(latexboundingbox(t, halign=halign, valign=valign)[1][2];digits=2) == 45.98
        @show round(latexboundingbox(t, halign=halign, valign=valign)[2][2];digits=2) == 0.0
    elseif valign === :middle
        @show round(latexboundingbox(t, halign=halign, valign=valign)[1][2];digits=2) == 45.98/2
        @show round(latexboundingbox(t, halign=halign, valign=valign)[2][2];digits=2) == -45.98/2
    end
end

finish()
d

rawlatexboundingbox(t)
```

```julia
d = Drawing(300,300,:svg)
origin()
fontsize(18)
t = L"x^2 + y^2"
halign = :center
valign = :middle
angle = -pi/2.5
# angle = 0 
pt = Point(20,20)
# pt = O
@layer begin
    translate(pt)
    text(t, halign=halign, valign=valign, angle=angle, rotationfixed=true)
end
@layer begin
    translate(pt)
    rotate(angle)
    box(latexboundingbox(t, halign=halign, valign=valign)...,:stroke)
end
circle(pt,2,:fill)
rulers()
finish()
d
```
